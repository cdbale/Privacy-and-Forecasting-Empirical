```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# nice time series plots
from sktime.utils.plotting import plot_series
```

```python
# import weekly finance time series
Y = pd.read_csv("../../Data/Results/DP_results.csv")
```

```python
LGBM_1 = Y.iloc[:,0]
LGBM_20 = Y.iloc[:,1]
ES_1 = Y.iloc[:,2]
ES_20 = Y.iloc[:,3]
Eps = Y.iloc[:,4]
```

```python
LGBM_1.index = [1, 10, 20]
LGBM_20.index = [1, 10, 20]
ES_1.index = [1, 10, 20]
ES_20.index = [1, 10, 20]
```

```python
LGBM_1
```

```python
plot_series(LGBM_1, LGBM_20, ES_1, ES_20, labels = ["LGBM, h=1", "LGBM, h=20", "SES, h=1", "SES, h=20"])
```

```python
plt.plot(LGBM_1, linestyle='--', marker='o', label="LGBM, h=1")
plt.plot(ES_1, linestyle='--', marker='o', label="SES, h=1")
plt.plot(LGBM_20, linestyle='--', marker='o', label="LGBM, h=20")
plt.plot(ES_20, linestyle='--', marker='o', label="SES, h=20")
plt.xticks([1, 10, 20])
plt.ylim([0, 600])
plt.xlabel("Epsilon")
plt.ylabel("MAE")
plt.title("MAE For LGBM vs. SES Under Differential Privacy")
plt.legend()
plt.rc('axes', labelsize=17)
plt.rc('xtick', labelsize=15)    # fontsize of the tick labels
plt.rc('ytick', labelsize=15)    # fontsize of the tick labels
plt.rc('legend', fontsize=15)    # legend fontsize
plt.grid()
plt.show()

plt.rcParams["figure.figsize"] = (13,10)
#fig.savefig('test2png.png', dpi=100)
```
